```python
import matplotlib.pyplot as plt

# c d liu t file
with open('../../data/runtime.txt', 'r') as file:
    lines = file.readlines()

# Dng u tin l tiu  cha cc gi tr k
k_values = list(lines[0].strip().split())

# Danh sch tn cho cc b d liu
dataset_names = ["Foodmart", "Retail", "Chess"]

# Duyt qua cc dng d liu, mi b gm 2 dng
for idx in range(1, len(lines), 2):
    # Ly d liu cho b hin ti
    runtime_TUHUFP = list(map(int, lines[idx].strip().split()))
    runtime_TUHUFP_Growth = list(map(int, lines[idx+1].strip().split()))

    # To mt biu  mi
    plt.figure(figsize=(10, 6))
    plt.plot(k_values, runtime_TUHUFP, marker='o', label=f'TUHUFP {dataset_names[idx//2]}', linestyle='--', linewidth=3)
    plt.plot(k_values, runtime_TUHUFP_Growth, marker='o', label=f'TUHUFP-Growth {dataset_names[idx//2]}', linewidth=3)

    # Thm cc ng k  nn
    plt.grid(True, axis='y')

    # Tiu  v nhn
    plt.title(f'Runtime Comparison of TUHUFP and TUHUFP-Growth - {dataset_names[idx//2]}')
    plt.xlabel('k')
    plt.ylabel('Runtime (seconds)')

    # Hin th legend
    plt.legend()

    # Hin th biu 
    plt.show()
```
